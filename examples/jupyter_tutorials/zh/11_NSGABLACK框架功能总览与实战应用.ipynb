## 实战应用1：生产调度优化

使用NSGABLACK框架解决一个实际的生产调度问题

```python
# 导入框架模块
try:
    from core.problems import Problem
    from core.solver import Solver
    from solvers.nsga2 import NSGA2
    from solvers.moead import MOEAD
    from bias.bias_v2 import BiasSystemV2
    from bias.bias_library_algorithmic import *
    from utils.experiment import Experiment
    from utils.parallel_evaluator import ParallelEvaluator
    print("框架模块导入成功")
except ImportError as e:
    print(f"模块导入失败: {e}")
    print("将使用简化实现进行演示")
    
    # 简化实现
    from abc import ABC, abstractmethod
    import random
    
    class Problem(ABC):
        def __init__(self, name, n_var, bounds):
            self.name = name
            self.n_var = n_var
            self.bounds = bounds
        
        @abstractmethod
        def evaluate(self, x):
            pass
    
    # 简化的调度器实现
    class ProductionScheduler:
        def __init__(self):
            pass
        
        def solve(self, jobs, machines):
            # 简化的调度算法
            schedule = []
            for job in jobs:
                machine = random.choice(machines)
                start_time = random.uniform(0, 100)
                schedule.append({
                    'job_id': job['id'],
                    'machine_id': machine,
                    'start_time': start_time,
                    'duration': job['duration']
                })
            return schedule
    
    print("使用简化实现进行演示")
```

```python
# 定义生产调度问题
class ProductionSchedulingProblem(Problem):
    """生产调度优化问题
    
    目标：
    1. 最小化总完成时间（makespan）
    2. 最小化总延迟
    3. 最大化机器利用率
    """
    
    def __init__(self, jobs, machines):
        self.jobs = jobs
        self.machines = machines
        n_var = len(jobs)  # 每个作业分配一个优先级
        bounds = [(0, 1)] * n_var
        super().__init__("ProductionScheduling", n_var, bounds)
        self.n_objectives = 3
    
    def evaluate(self, x):
        """评估调度方案"""
        # 使用优先级生成调度
        schedule = self._generate_schedule(x)
        
        # 计算目标值
        makespan = self._calculate_makespan(schedule)
        total_tardiness = self._calculate_total_tardiness(schedule)
        machine_utilization = self._calculate_machine_utilization(schedule)
        
        # 返回加权总目标（简化处理）
        total_objective = 0.4 * makespan + 0.3 * total_tardiness - 0.3 * machine_utilization
        
        return total_objective
    
    def evaluate_objectives(self, x):
        """评估所有目标（用于多目标优化）"""
        schedule = self._generate_schedule(x)
        
        makespan = self._calculate_makespan(schedule)
        total_tardiness = self._calculate_total_tardiness(schedule)
        machine_utilization = self._calculate_machine_utilization(schedule)
        
        return np.array([makespan, total_tardiness, -machine_utilization])
    
    def _generate_schedule(self, priorities):
        """根据优先级生成调度"""
        # 按优先级排序作业
        job_priorities = [(i, priorities[i], self.jobs[i]) 
                          for i in range(len(self.jobs))]
        job_priorities.sort(key=lambda x: x[1], reverse=True)
        
        schedule = []
        machine_availability = {m['id']: 0 for m in self.machines}
        
        for job_id, priority, job in job_priorities:
            # 选择最早可用的机器
            best_machine = min(machine_availability.items(), key=lambda x: x[1])[0]
            
            start_time = machine_availability[best_machine]
            
            schedule.append({
                'job_id': job['id'],
                'machine_id': best_machine,
                'start_time': start_time,
                'duration': job['duration'],
                'end_time': start_time + job['duration'],
                'priority': priority
            })
            
            machine_availability[best_machine] = start_time + job['duration']
        
        return schedule
    
    def _calculate_makespan(self, schedule):
        """计算总完成时间"""
        if not schedule:
            return 0
        return max(job['end_time'] for job in schedule)
    
    def _calculate_total_tardiness(self, schedule):
        """计算总延迟"""
        total_tardiness = 0
        for job_info in schedule:
            job = next(j for j in self.jobs if j['id'] == job_info['job_id'])
            due_date = job.get('due_date', float('inf'))
            tardiness = max(0, job_info['end_time'] - due_date)
            total_tardiness += tardiness
        return total_tardiness
    
    def _calculate_machine_utilization(self, schedule):
        """计算机器利用率"""
        if not schedule:
            return 0
        
        total_processing_time = sum(job['duration'] for job in schedule)
        makespan = self._calculate_makespan(schedule)
        n_machines = len(self.machines)
        
        if makespan > 0:
            return total_processing_time / (makespan * n_machines)
        return 0

# 创建测试数据
def create_test_problem():
    """创建测试调度问题"""
    # 定义作业
    jobs = [
        {'id': 1, 'duration': 10, 'due_date': 15},
        {'id': 2, 'duration': 8, 'due_date': 20},
        {'id': 3, 'duration': 12, 'due_date': 25},
        {'id': 4, 'duration': 6, 'due_date': 18},
        {'id': 5, 'duration': 15, 'due_date': 30},
        {'id': 6, 'duration': 9, 'due_date': 22},
        {'id': 7, 'duration': 11, 'due_date': 28},
        {'id': 8, 'duration': 7, 'due_date': 16},
    ]
    
    # 定义机器
    machines = [
        {'id': 'M1', 'capacity': 1},
        {'id': 'M2', 'capacity': 1},
        {'id': 'M3', 'capacity': 1},
    ]
    
    return ProductionSchedulingProblem(jobs, machines), jobs, machines

# 创建问题实例
problem, jobs, machines = create_test_problem()

print(f"生产调度问题实例:")
print(f"  作业数: {len(jobs)}")
print(f"  机器数: {len(machines)}")
print(f"  决策变量数: {problem.n_var}")
print(f"  目标数: {problem.n_objectives}")
```

## 实战应用2：使用NSGA-II进行多目标优化

```python
# 简化的NSGA-II实现（如果框架模块不可用）
class SimplifiedNSGA2:
    """简化的NSGA-II实现"""
    
    def __init__(self, population_size=100, max_generations=100, 
                 crossover_rate=0.9, mutation_rate=0.1):
        self.population_size = population_size
        self.max_generations = max_generations
        self.crossover_rate = crossover_rate
        self.mutation_rate = mutation_rate
        
        self.pareto_front = []
        self.history = []
    
    def optimize(self, problem):
        """执行优化"""
        print(f"\n开始NSGA-II优化...")
        print(f"种群大小: {self.population_size}")
        print(f"最大代数: {self.max_generations}")
        
        # 初始化种群
        population = self._initialize_population(problem)
        
        for generation in range(self.max_generations):
            # 评估种群
            objectives = []
            for individual in population:
                obj = problem.evaluate_objectives(individual)
                objectives.append(obj)
            
            # 非支配排序
            fronts = self._fast_non_dominated_sort(objectives)
            
            # 更新帕累托前沿
            if fronts:
                self.pareto_front = [population[i] for i in fronts[0]]
                self.history.append(objectives)
            
            # 输出进度
            if (generation + 1) % 20 == 0:
                print(f"  代数 {generation+1}: 前沿大小 = {len(self.pareto_front)}")
            
            # 生成新一代
            offspring = self._generate_offspring(population, problem)
            
            # 合并和选择
            population = self._select_next_generation(population, offspring, problem)
        
        return self.pareto_front
    
    def _initialize_population(self, problem):
        """初始化种群"""
        population = []
        for _ in range(self.population_size):
            individual = np.array([
                np.random.uniform(low, high) 
                for low, high in problem.bounds
            ])
            population.append(individual)
        return population
    
    def _fast_non_dominated_sort(self, objectives):
        """快速非支配排序"""
        n = len(objectives)
        domination_count = [0] * n
        dominated_sets = [[] for _ in range(n)]
        fronts = [[]]
        
        # 计算支配关系
        for i in range(n):
            for j in range(n):
                if i != j:
                    if self._dominates(objectives[i], objectives[j]):
                        dominated_sets[i].append(j)
                        domination_count[j] += 1
        
        # 找到第一前沿
        for i in range(n):
            if domination_count[i] == 0:
                fronts[0].append(i)
        
        # 找到后续前沿
        i = 0
        while fronts[i]:
            next_front = []
            for individual in fronts[i]:
                for dominated in dominated_sets[individual]:
                    domination_count[dominated] -= 1
                    if domination_count[dominated] == 0:
                        next_front.append(dominated)
            i += 1
            fronts.append(next_front)
        
        return fronts[:-1]  # 移除最后的空前沿
    
    def _dominates(self, obj1, obj2):
        """检查obj1是否支配obj2"""
        return np.all(obj1 <= obj2) and np.any(obj1 < obj2)
    
    def _generate_offspring(self, population, problem):
        """生成后代"""
        offspring = []
        
        while len(offspring) < self.population_size:
            # 选择父代（锦标赛选择）
            parent1 = self._tournament_selection(population, problem)
            parent2 = self._tournament_selection(population, problem)
            
            # 交叉
            if np.random.random() < self.crossover_rate:
                child1, child2 = self._crossover(parent1, parent2)
            else:
                child1, child2 = parent1.copy(), parent2.copy()
            
            # 变异
            if np.random.random() < self.mutation_rate:
                child1 = self._mutate(child1, problem)
            if np.random.random() < self.mutation_rate:
                child2 = self._mutate(child2, problem)
            
            offspring.append(child1)
            if len(offspring) < self.population_size:
                offspring.append(child2)
        
        return offspring[:self.population_size]
    
    def _tournament_selection(self, population, problem, tournament_size=3):
        """锦标赛选择"""
        candidates = np.random.choice(len(population), tournament_size, replace=False)
        best = candidates[0]
        best_obj = problem.evaluate_objectives(population[best])
        
        for candidate in candidates[1:]:
            obj = problem.evaluate_objectives(population[candidate])
            if self._dominates(obj, best_obj):
                best = candidate
                best_obj = obj
        
        return population[best].copy()
    
    def _crossover(self, parent1, parent2):
        """模拟二进制交叉"""
        child1 = parent1.copy()
        child2 = parent2.copy()
        
        for i in range(len(parent1)):
            if np.random.random() < 0.5:
                child1[i], child2[i] = child2[i], child1[i]
        
        return child1, child2
    
    def _mutate(self, individual, problem):
        """多项式变异"""
        mutated = individual.copy()
        for i in range(len(mutated)):
            if np.random.random() < 1.0 / len(mutated):
                low, high = problem.bounds[i]
                mutated[i] = np.random.uniform(low, high)
        return mutated
    
    def _select_next_generation(self, population, offspring, problem):
        """选择下一代"""
        combined = population + offspring
        
        # 评估合并种群
        objectives = []
        for individual in combined:
            obj = problem.evaluate_objectives(individual)
            objectives.append(obj)
        
        # 非支配排序
        fronts = self._fast_non_dominated_sort(objectives)
        
        # 选择下一代
        next_generation = []
        for front in fronts:
            if len(next_generation) + len(front) <= self.population_size:
                next_generation.extend([combined[i] for i in front])
            else:
                # 使用拥挤度选择
                remaining = self.population_size - len(next_generation)
                front_objectives = [objectives[i] for i in front]
                crowding_distances = self._calculate_crowding_distance(front_objectives)
                
                # 按拥挤度排序
                sorted_indices = np.argsort(crowding_distances)[::-1]
                for idx in sorted_indices[:remaining]:
                    next_generation.append(combined[front[idx]])
                break
        
        return next_generation
    
    def _calculate_crowding_distance(self, objectives):
        """计算拥挤度距离"""
        objectives = np.array(objectives)
        n_points, n_obj = objectives.shape
        crowding = np.zeros(n_points)
        
        for i in range(n_obj):
            sorted_indices = np.argsort(objectives[:, i])
            obj_range = objectives[sorted_indices[-1], i] - objectives[sorted_indices[0], i]
            
            if obj_range > 0:
                for j in range(1, n_points - 1):
                    crowding[sorted_indices[j]] += (
                        objectives[sorted_indices[j+1], i] - objectives[sorted_indices[j-1], i]
                    ) / obj_range
        
        return crowding

# 运行多目标优化
print("\n运行NSGA-II多目标优化...\n" + "="*50)

# 创建优化器
try:
    # 尝试使用框架中的NSGA-II
    nsga2 = NSGA2(
        population_size=50,
        max_generations=100,
        crossover_rate=0.9,
        mutation_rate=0.1
    )
except:
    # 使用简化实现
    nsga2 = SimplifiedNSGA2(
        population_size=50,
        max_generations=100,
        crossover_rate=0.9,
        mutation_rate=0.1
    )

# 执行优化
start_time = time.time()
pareto_solutions = nsga2.optimize(problem)
elapsed_time = time.time() - start_time

print(f"\n优化完成！")
print(f"运行时间: {elapsed_time:.2f}秒")
print(f"帕累托前沿大小: {len(pareto_solutions)}")

# 评估帕累托前沿
if pareto_solutions:
    pareto_objectives = []
    for solution in pareto_solutions:
        obj = problem.evaluate_objectives(solution)
        pareto_objectives.append(obj)
    pareto_objectives = np.array(pareto_objectives)
    
    print(f"\n帕累托前沿统计：")
    print(f"  最小总完成时间: {np.min(pareto_objectives[:, 0]):.2f}")
    print(f"  最小总延迟: {np.min(pareto_objectives[:, 1]):.2f}")
    print(f"  最大机器利用率: {np.max(-pareto_objectives[:, 2]):.2f}")
    
    # 可视化结果
    plt.figure(figsize=(15, 5))
    
    # 帕累托前沿
    plt.subplot(1, 3, 1)
    plt.scatter(pareto_objectives[:, 0], pareto_objectives[:, 1], 
               c=-pareto_objectives[:, 2], cmap='viridis', s=50)
    plt.xlabel('总完成时间')
    plt.ylabel('总延迟')
    plt.title('帕累托前沿')
    plt.colorbar(label='机器利用率')
    plt.grid(True, alpha=0.3)
    
    # 目标分布
    plt.subplot(1, 3, 2)
    objectives_labels = ['总完成时间', '总延迟', '机器利用率']
    for i in range(3):
        plt.subplot(2, 2, i+1)
        plt.hist(pareto_objectives[:, i], bins=20, alpha=0.7)
        plt.xlabel(objectives_labels[i])
        plt.ylabel('频次')
        plt.title(f'{objectives_labels[i]}分布')
        plt.grid(True, alpha=0.3)
    
    # 3D散点图
    from mpl_toolkits.mplot3d import Axes3D
    ax = plt.subplot(1, 3, 3, projection='3d')
    scatter = ax.scatter(pareto_objectives[:, 0], 
                         pareto_objectives[:, 1], 
                         -pareto_objectives[:, 2],
                         c=range(len(pareto_objectives)), cmap='viridis')
    ax.set_xlabel('总完成时间')
    ax.set_ylabel('总延迟')
    ax.set_zlabel('机器利用率')
    ax.set_title('三维目标空间')
    
    plt.tight_layout()
    plt.show()
    
    # 选择几个代表性解
    print("\n代表性调度方案：")
    print("-" * 50)
    
    # 1. 最小化总完成时间
    min_makespan_idx = np.argmin(pareto_objectives[:, 0])
    best_makespan = pareto_solutions[min_makespan_idx]
    schedule_makespan = problem._generate_schedule(best_makespan)
    
    print("\n方案1 - 最小总完成时间:")
    for job in schedule_makespan:
        print(f"  作业{job['job_id']}: 机器{job['machine_id']}, "
              f"开始时间={job['start_time']:.1f}, 结束时间={job['end_time']:.1f}")
    print(f"  总完成时间: {problem._calculate_makespan(schedule_makespan):.1f}")
    
    # 2. 最小化总延迟
    min_tardiness_idx = np.argmin(pareto_objectives[:, 1])
    best_tardiness = pareto_solutions[min_tardiness_idx]
    schedule_tardiness = problem._generate_schedule(best_tardiness)
    
    print("\n方案2 - 最小总延迟:")
    for job in schedule_tardiness:
        print(f"  作业{job['job_id']}: 机器{job['machine_id']}, "
              f"开始时间={job['start_time']:.1f}, 结束时间={job['end_time']:.1f}")
    print(f"  总延迟: {problem._calculate_total_tardiness(schedule_tardiness):.1f}")
    
    # 3. 最大化机器利用率
    max_utilization_idx = np.argmax(-pareto_objectives[:, 2])
    best_utilization = pareto_solutions[max_utilization_idx]
    schedule_utilization = problem._generate_schedule(best_utilization)
    
    print("\n方案3 - 最高机器利用率:")
    for job in schedule_utilization:
        print(f"  作业{job['job_id']}: 机器{job['machine_id']}, "
              f"开始时间={job['start_time']:.1f}, 结束时间={job['end_time']:.1f}")
    print(f"  机器利用率: {problem._calculate_machine_utilization(schedule_utilization):.2%}")
else:
    print("未找到可行解")
```

## 实战应用3：使用偏置系统增强优化

```python
# 简化的偏置系统实现
class ProductionSchedulingBias:
    """生产调度偏置 - 引入领域知识"""
    
    def __init__(self, due_date_weight=0.5, processing_time_weight=0.5):
        self.due_date_weight = due_date_weight
        self.processing_time_weight = processing_time_weight
    
    def apply(self, population, problem):
        """应用偏置到种群"""
        biased_population = []
        
        for individual in population:
            # 基于截止日期和加工时间的启发式偏置
            biased_individual = self._apply_heuristic_bias(individual, problem)
            biased_population.append(biased_individual)
        
        return biased_population
    
    def _apply_heuristic_bias(self, individual, problem):
        """应用启发式偏置"""
        biased = individual.copy()
        
        # 计算每个作业的优先级分数
        priorities = []
        for i, job in enumerate(problem.jobs):
            due_date_score = 1.0 / max(1, job['due_date'])
            processing_score = 1.0 / max(1, job['duration'])
            
            priority = (self.due_date_weight * due_date_score + 
                       self.processing_time_weight * processing_score)
            priorities.append(priority)
        
        # 归一化优先级
        priorities = np.array(priorities)
        priorities = (priorities - priorities.min()) / (priorities.max() - priorities.min() + 1e-10)
        
        # 应用偏置：部分个体采用启发式优先级
        if np.random.random() < 0.3:  # 30%的个体应用偏置
            noise_factor = 0.1
            biased = priorities + np.random.normal(0, noise_factor, len(priorities))
            biased = np.clip(biased, 0, 1)
        
        return biased

# 带偏置的遗传算法
class BiasedGA:
    """带偏置的遗传算法"""
    
    def __init__(self, population_size=50, max_generations=100, bias=None):
        self.population_size = population_size
        self.max_generations = max_generations
        self.bias = bias
        
        self.best_solution = None
        self.best_fitness = float('inf')
        self.history = []
    
    def optimize(self, problem):
        """执行优化"""
        print(f"\n运行带偏置的遗传算法...")
        print(f"偏置类型: {type(self.bias).__name__ if self.bias else '无'}")
        
        # 初始化种群
        population = self._initialize_population(problem)
        
        for generation in range(self.max_generations):
            # 应用偏置
            if self.bias:
                population = self.bias.apply(population, problem)
            
            # 评估种群
            fitnesses = []
            for individual in population:
                fitness = problem.evaluate(individual)
                fitnesses.append(fitness)
            
            # 更新最优解
            min_idx = np.argmin(fitnesses)
            if fitnesses[min_idx] < self.best_fitness:
                self.best_fitness = fitnesses[min_idx]
                self.best_solution = population[min_idx].copy()
            
            self.history.append(self.best_fitness)
            
            # 输出进度
            if (generation + 1) % 20 == 0:
                print(f"  代数 {generation+1}: 最优值 = {self.best_fitness:.4f}")
            
            # 选择、交叉、变异
            population = self._evolve_population(population, fitnesses, problem)
        
        return self.best_solution, self.best_fitness
    
    def _initialize_population(self, problem):
        """初始化种群"""
        population = []
        for _ in range(self.population_size):
            individual = np.array([
                np.random.uniform(low, high) 
                for low, high in problem.bounds
            ])
            population.append(individual)
        return population
    
    def _evolve_population(self, population, fitnesses, problem):
        """进化种群"""
        # 锦标赛选择
        selected = []
        for _ in range(self.population_size):
            tournament_size = 3
            candidates = np.random.choice(len(population), tournament_size, replace=False)
            best = candidates[0]
            for candidate in candidates[1:]:
                if fitnesses[candidate] < fitnesses[best]:
                    best = candidate
            selected.append(population[best].copy())
        
        # 交叉和变异
        new_population = []
        for i in range(0, len(selected), 2):
            parent1 = selected[i]
            parent2 = selected[i+1] if i+1 < len(selected) else selected[0]
            
            # 交叉
            if np.random.random() < 0.8:
                child1, child2 = self._crossover(parent1, parent2)
            else:
                child1, child2 = parent1.copy(), parent2.copy()
            
            # 变异
            child1 = self._mutate(child1, problem)
            child2 = self._mutate(child2, problem)
            
            new_population.extend([child1, child2])
        
        return new_population[:self.population_size]
    
    def _crossover(self, parent1, parent2):
        """交叉操作"""
        child1 = parent1.copy()
        child2 = parent2.copy()
        
        # 单点交叉
        crossover_point = np.random.randint(1, len(parent1))
        child1[crossover_point:] = parent2[crossover_point:]
        child2[crossover_point:] = parent1[crossover_point:]
        
        return child1, child2
    
    def _mutate(self, individual, problem):
        """变异操作"""
        mutated = individual.copy()
        mutation_rate = 1.0 / len(mutated)
        
        for i in range(len(mutated)):
            if np.random.random() < mutation_rate:
                low, high = problem.bounds[i]
                mutated[i] = np.random.uniform(low, high)
        
        return mutated

# 对比实验：无偏置 vs 有偏置
print("\n对比实验：无偏置 vs 有偏置\n" + "="*50)

# 无偏置的GA
ga_no_bias = BiasedGA(population_size=50, max_generations=100, bias=None)
best_no_bias, fitness_no_bias = ga_no_bias.optimize(problem)

# 有偏置的GA
bias = ProductionSchedulingBias(due_date_weight=0.7, processing_time_weight=0.3)
ga_with_bias = BiasedGA(population_size=50, max_generations=100, bias=bias)
best_with_bias, fitness_with_bias = ga_with_bias.optimize(problem)

# 比较结果
print(f"\n结果对比：")
print(f"无偏置GA:")
print(f"  最优适应度: {fitness_no_bias:.4f}")
print(f"  最优解: {best_no_bias}")

print(f"\n有偏置GA:")
print(f"  最优适应度: {fitness_with_bias:.4f}")
print(f"  最优解: {best_with_bias}")

improvement = (fitness_no_bias - fitness_with_bias) / fitness_no_bias * 100
print(f"\n改进: {improvement:.2f}%")

# 可视化收敛曲线
plt.figure(figsize=(10, 6))

plt.plot(ga_no_bias.history, 'b-', label='无偏置GA', linewidth=2)
plt.plot(ga_with_bias.history, 'r-', label='有偏置GA', linewidth=2)
plt.xlabel('代数')
plt.ylabel('最优适应度')
plt.title('收敛曲线对比')
plt.legend()
plt.grid(True, alpha=0.3)
plt.yscale('log')

plt.show()
```

## 实战应用4：并行优化加速

```python
# 简化的并行优化实现
class ParallelOptimization:
    """并行优化框架"""
    
    def __init__(self, n_workers=4):
        self.n_workers = n_workers
        self.results = []
    
    def parallel_optimize(self, problem, solver_class, 
                           solver_configs, max_runs=10):
        """并行执行多次优化"""
        import concurrent.futures
        
        print(f"\n启动并行优化...")
        print(f"工作进程数: {self.n_workers}")
        print(f"配置数量: {len(solver_configs)}")
        print(f"每配置运行次数: {max_runs}")
        
        def run_optimization(config_id, config, run_id):
            """单次优化运行"""
            # 创建求解器实例
            solver = solver_class(**config)
            
            # 执行优化
            start_time = time.time()
            result = solver.optimize(problem)
            elapsed_time = time.time() - start_time
            
            return {
                'config_id': config_id,
                'config': config,
                'run_id': run_id,
                'result': result,
                'elapsed_time': elapsed_time
            }
        
        # 生成所有任务
        tasks = []
        for config_id, config in enumerate(solver_configs):
            for run_id in range(max_runs):
                tasks.append((config_id, config, run_id))
        
        # 并行执行
        all_results = []
        with concurrent.futures.ProcessPoolExecutor(max_workers=self.n_workers) as executor:
            # 提交所有任务
            futures = [
                executor.submit(run_optimization, *task)
                for task in tasks
            ]
            
            # 收集结果
            for future in concurrent.futures.as_completed(futures):
                try:
                    result = future.result()
                    all_results.append(result)
                    print(f"  完成配置{result['config_id']}, 运行{result['run_id']}")
                except Exception as e:
                    print(f"  错误: {e}")
        
        self.results = all_results
        return all_results
    
    def analyze_results(self):
        """分析并行优化结果"""
        if not self.results:
            print("没有结果可分析")
            return
        
        # 按配置分组
        config_results = {}
        for result in self.results:
            config_id = result['config_id']
            if config_id not in config_results:
                config_results[config_id] = {
                    'config': result['config'],
                    'results': [],
                    'best_fitness': float('inf'),
                    'worst_fitness': 0,
                    'mean_fitness': 0,
                    'mean_time': 0
                }
            
            config_results[config_id]['results'].append(result)
        
        # 计算统计信息
        for config_id, group in config_results.items():
            fitnesses = [r['result'].get('best_fitness', float('inf')) 
                         for r in group['results']]
            times = [r['elapsed_time'] for r in group['results']]
            
            group['best_fitness'] = min(fitnesses)
            group['worst_fitness'] = max(fitnesses)
            group['mean_fitness'] = np.mean(fitnesses)
            group['std_fitness'] = np.std(fitnesses)
            group['mean_time'] = np.mean(times)
        
        # 找到最佳配置
        best_config_id = min(config_results.keys(), 
                            key=lambda x: config_results[x]['best_fitness'])
        best_config = config_results[best_config_id]
        
        print(f"\n并行优化结果分析：")
        print(f"="*50)
        
        for config_id, group in config_results.items():
            print(f"\n配置 {config_id}:")
            print(f"  参数: {group['config']}")
            print(f"  最优适应度: {group['best_fitness']:.4f}")
            print(f"  平均适应度: {group['mean_fitness']:.4f} ± {group['std_fitness']:.4f}")
            print(f"  平均运行时间: {group['mean_time']:.2f}秒")
        
        print(f"\n最佳配置: {best_config_id}")
        print(f"  参数: {best_config['config']}")
        print(f"  最优适应度: {best_config['best_fitness']:.4f}")
        
        return config_results

# 定义不同的求解器配置
class ConfigurableGA:
    """可配置的遗传算法"""
    
    def __init__(self, population_size=50, max_generations=100,
                 crossover_rate=0.8, mutation_rate=0.1):
        self.population_size = population_size
        self.max_generations = max_generations
        self.crossover_rate = crossover_rate
        self.mutation_rate = mutation_rate
    
    def optimize(self, problem):
        """简化优化"""
        best_fitness = float('inf')
        
        for _ in range(self.max_generations):
            # 简单的随机搜索模拟
            x = np.array([np.random.uniform(low, high) 
                           for low, high in problem.bounds])
            fitness = problem.evaluate(x)
            
            if fitness < best_fitness:
                best_fitness = fitness
        
        return {'best_fitness': best_fitness}

# 配置不同的算法参数
solver_configs = [
    {'population_size': 30, 'max_generations': 50, 
     'crossover_rate': 0.7, 'mutation_rate': 0.2},
    {'population_size': 50, 'max_generations': 100, 
     'crossover_rate': 0.8, 'mutation_rate': 0.1},
    {'population_size': 100, 'max_generations': 100, 
     'crossover_rate': 0.9, 'mutation_rate': 0.05},
    {'population_size': 50, 'max_generations': 200, 
     'crossover_rate': 0.8, 'mutation_rate': 0.1},
]

# 执行并行优化
parallel_opt = ParallelOptimization(n_workers=2)  # 使用2个工作进程
results = parallel_opt.parallel_optimize(
    problem,
    ConfigurableGA,
    solver_configs,
    max_runs=5
)

# 分析结果
analysis = parallel_opt.analyze_results()

# 可视化结果
if analysis:
    plt.figure(figsize=(12, 5))
    
    # 性能对比
    plt.subplot(1, 2, 1)
    config_ids = list(analysis.keys())
    best_fitnesses = [analysis[cid]['best_fitness'] for cid in config_ids]
    mean_fitnesses = [analysis[cid]['mean_fitness'] for cid in config_ids]
    std_fitnesses = [analysis[cid]['std_fitness'] for cid in config_ids]
    
    x_pos = np.arange(len(config_ids))
    plt.bar(x_pos, best_fitnesses, yerr=std_fitnesses, 
            capsize=5, alpha=0.7, label='最优±标准差')
    plt.plot(x_pos, mean_fitnesses, 'ro-', label='平均值')
    plt.xlabel('配置ID')
    plt.ylabel('适应度')
    plt.title('不同配置的性能对比')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 运行时间对比
    plt.subplot(1, 2, 2)
    mean_times = [analysis[cid]['mean_time'] for cid in config_ids]
    bars = plt.bar(x_pos, mean_times, alpha=0.7, color='orange')
    plt.xlabel('配置ID')
    plt.ylabel('平均运行时间（秒）')
    plt.title('不同配置的运行时间')
    plt.grid(True, alpha=0.3, axis='y')
    
    # 添加数值标签
    for bar, time_val in zip(bars, mean_times):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), 
                 f'{time_val:.2f}s', ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()
```
