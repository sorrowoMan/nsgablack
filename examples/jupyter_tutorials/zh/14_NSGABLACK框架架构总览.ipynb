## 2. 核心接口定义

展示框架中的核心接口和抽象类定义。

```python
# 模拟核心接口定义
from abc import ABC, abstractmethod
from typing import Protocol, runtime_checkable

@runtime_checkable
class ProblemInterface(Protocol):
    """问题接口协议"""
    
    @property
    def name(self) -> str:
        """问题名称"""
        ...
    
    @property
    def n_var(self) -> int:
        """变量维度"""
        ...
    
    @property
    def bounds(self) -> List[Tuple[float, float]]:
        """变量边界"""
        ...
    
    def evaluate(self, x: np.ndarray) -> float:
        """评估单个解"""
        ...

class SolverInterface(ABC):
    """求解器接口抽象类"""
    
    def __init__(self, name: str, **kwargs):
        self.name = name
        self.parameters = kwargs
        self.status = 'idle'
        self.best_solution = None
        self.best_fitness = float('inf')
        self.history = []
    
    @abstractmethod
    def _initialize(self, problem: 'ProblemInterface') -> None:
        """初始化求解器"""
        pass
    
    @abstractmethod
    def _step(self, problem: 'ProblemInterface') -> None:
        """执行一步优化"""
        pass
    
    def solve(self, problem: 'ProblemInterface', 
              max_evaluations: int = 10000, 
              **kwargs) -> Dict[str, Any]:
        """标准求解流程"""
        # 模板方法模式
        self._pre_solve(problem, **kwargs)
        
        # 初始化
        self._initialize(problem)
        self.status = 'running'
        
        # 主循环
        self._optimize_loop(problem, max_evaluations, **kwargs)
        
        # 后处理
        self._post_solve(**kwargs)
        
        return self.get_results()
    
    def _pre_solve(self, problem, **kwargs):
        """求解前准备"""
        print(f"开始求解: {problem.name}")
        print(f"使用算法: {self.name}")
    
    def _optimize_loop(self, problem, max_evaluations, **kwargs):
        """优化主循环"""
        n_evaluations = 0
        
        while n_evaluations < max_evaluations and self.status == 'running':
            self._step(problem)
            n_evaluations += 1
            
            # 记录历史
            self._record_iteration(n_evaluations)
            
            # 检查停止条件
            if self._check_stop_condition(**kwargs):
                break
    
    def _post_solve(self, **kwargs):
        """求解后处理"""
        self.status = 'finished'
        print(f"求解完成，最优适应度: {self.best_fitness}")
    
    def _record_iteration(self, n_evaluations: int):
        """记录迭代信息"""
        self.history.append({
            'evaluations': n_evaluations,
            'best_fitness': self.best_fitness,
            'solution': self.best_solution.copy() if self.best_solution is not None else None
        })
    
    def _check_stop_condition(self, **kwargs) -> bool:
        """检查停止条件"""
        # 子类可以重写此方法
        return False
    
    def get_results(self) -> Dict[str, Any]:
        """获取求解结果"""
        return {
            'best_solution': self.best_solution,
            'best_fitness': self.best_fitness,
            'history': self.history.copy(),
            'parameters': self.parameters
        }

class BiasInterface(ABC):
    """偏置接口抽象类"""
    
    def __init__(self, name: str, strength: float = 1.0):
        self.name = name
        self.strength = strength
        self.enabled = True
    
    @abstractmethod
    def apply(self, population: List[np.ndarray], 
               problem: 'ProblemInterface') -> List[np.ndarray]:
        """应用偏置到种群"""
        pass
    
    def set_strength(self, strength: float):
        """设置偏置强度"""
        self.strength = max(0.0, min(1.0, strength))
    
    def enable(self):
        """启用偏置"""
        self.enabled = True
    
    def disable(self):
        """禁用偏置"""
        self.enabled = False

class EvaluatorInterface(Protocol):
    """评估器接口协议"""
    
    def evaluate(self, func: Callable, 
                  candidates: List[Any]) -> List[Any]:
        """评估候选解"""
        ...

# 展示接口的多态性
print("\n核心接口定义：")
print("=" * 50)

print("\n1. 问题接口 (ProblemInterface)")
print("   - 标准化问题定义")
print("   - 支持单目标和多目标")
print("   - 统一的边界和约束处理")

print("\n2. 求解器接口 (SolverInterface)")
print("   - 模板方法定义求解流程")
print("   - 统一的结果和状态管理")
print("   - 支持回调钩子")

print("\n3. 偏置接口 (BiasInterface)")
print("   - 可插拔的偏置策略")
print("   - 动态强度调整")
print("   - 支持组合偏置")

print("\n4. 评估器接口 (EvaluatorInterface)")
print("   - 抽象评估过程")
print("   - 支持串行和并行")
print("   - 可扩展的评估策略")

# 展示接口实现示例
print("\n接口实现示例：")
print("-" * 30)

class ExampleSolver(SolverInterface):
    """示例求解器实现"""
    
    def __init__(self, **kwargs):
        super().__init__("ExampleSolver", **kwargs)
        self.population = []
    
    def _initialize(self, problem: ProblemInterface):
        # 初始化种群
        self.population = []
        for _ in range(self.parameters.get('population_size', 50)):
            x = np.array([np.random.uniform(low, high) 
                          for low, high in problem.bounds])
            self.population.append(x)
    
    def _step(self, problem: ProblemInterface):
        # 简单的随机搜索步骤
        for i in range(len(self.population)):
            if np.random.random() < 0.1:  # 10%的概率变异
                x = np.array([np.random.uniform(low, high) 
                              for low, high in problem.bounds])
                fitness = problem.evaluate(x)
                
                if fitness < self.best_fitness:
                    self.best_fitness = fitness
                    self.best_solution = x.copy()

print("\n创建示例求解器：")
solver = ExampleSolver(population_size=100, max_generations=500)
print(f"求解器名称: {solver.name}")
print(f"参数: {solver.parameters}")
print(f"状态: {solver.status}")
```

## 3. 模块间交互机制

展示各模块之间如何协同工作。

```python
class ModuleInteractionAnalyzer:
    """模块交互分析器"""
    
    def __init__(self):
        self.interaction_matrix = self._build_interaction_matrix()
        self.data_flow = self._analyze_data_flow()
        self.control_flow = self._analyze_control_flow()
    
    def _build_interaction_matrix(self):
        """构建模块交互矩阵"""
        modules = ['core', 'bias', 'solvers', 'utils']
        interaction_matrix = {}
        
        for module in modules:
            interaction_matrix[module] = {}
            for other_module in modules:
                interaction_matrix[module][other_module] = self._get_interaction_strength(
                    module, other_module
                )
        
        return interaction_matrix
    
    def _get_interaction_strength(self, module1: str, module2: str) -> int:
        """获取两个模块间的交互强度"""
        # 定义交互强度矩阵
        interactions = {
            ('core', 'bias'): 3,  # bias依赖core
            ('core', 'solvers'): 3,  # solvers依赖core
            ('core', 'utils'): 2,  # utils依赖core
            ('bias', 'core'): 1,  # core不依赖bias
            ('bias', 'solvers'): 2,  # solvers使用bias
            ('bias', 'utils'): 1,
            ('solvers', 'core'): 1,
            ('solvers', 'bias'): 1,
            ('solvers', 'utils'): 2,
            ('utils', 'core'): 1,
            ('utils', 'bias'): 1,
            ('utils', 'solvers'): 1,
        }
        
        if module1 == module2:
            return 5  # 模块内部交互最强
        
        return interactions.get((module1, module2), 0)
    
    def _analyze_data_flow(self):
        """分析数据流"""
        data_flow = {
            '问题定义': {
                'source': 'core/problems.py',
                'destinations': ['solvers/*', 'utils/parallel_evaluator.py'],
                'data_types': ['问题实例', '边界信息', '约束定义']
            },
            '优化解': {
                'source': 'solvers/*',
                'destinations': ['core/elite.py', 'bias/*'],
                'data_types': ['解向量', '适应度值', '目标值']
            },
            '偏置信息': {
                'source': 'bias/*',
                'destinations': ['solvers/*'],
                'data_types': ['偏置向量', '强度参数', '策略信息']
            },
            '性能指标': {
                'source': 'all modules',
                'destinations': ['utils/experiment.py'],
                'data_types': ['运行时间', '内存使用', '收敛信息']
            }
        }
        return data_flow
    
    def _analyze_control_flow(self):
        """分析控制流"""
        control_flow = {
            '主流程': [
                '用户调用 -> Solver.solve()',
                'Solver.solve() -> Problem.evaluate()',
                'Solver.solve() -> Bias.apply() (可选)',
                'Solver.solve() -> EliteManager.update()',
                '结果返回 -> 用户'
            ],
            '回调流程': [
                '优化事件 -> Observer.notify()',
                'Observer.notify() -> 回调函数执行',
                '回调函数 -> 日志记录/可视化'
            ],
            '并行流程': [
                '主进程 -> ParallelEvaluator',
                'ParallelEvaluator -> Worker进程',
                'Worker进程 -> Problem.evaluate()',
                '结果聚合 -> 主进程'
            ]
        }
        return control_flow
    
    def visualize_interactions(self):
        """可视化模块交互"""
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))
        
        # 1. 交互强度热图
        ax1 = axes[0, 0]
        modules = list(self.interaction_matrix.keys())
        matrix = np.array([[self.interaction_matrix[m][n] for n in modules] 
                          for m in modules])
        
        im = ax1.imshow(matrix, cmap='YlOrRd', vmin=0, vmax=5)
        ax1.set_xticks(range(len(modules)))
        ax1.set_yticks(range(len(modules)))
        ax1.set_xticklabels(modules)
        ax1.set_yticklabels(modules)
        ax1.set_title('模块交互强度')
        
        # 添加数值标签
        for i in range(len(modules)):
            for j in range(len(modules)):
                ax1.text(j, i, str(matrix[i, j]), ha='center', va='center')
        
        plt.colorbar(im, ax=ax1, label='交互强度')
        
        # 2. 数据流图
        ax2 = axes[0, 1]
        for flow_name, flow_info in self.data_flow.items():
            ax2.text(0.1, 0.9 - list(self.data_flow.keys()).index(flow_name) * 0.2,
                   f"{flow_name}:", fontweight='bold')
            ax2.text(0.1, 0.85 - list(self.data_flow.keys()).index(flow_name) * 0.2,
                   f"  {flow_info['source']} → {','.join(flow_info['destinations'][:2])}", 
                   fontsize=9)
        ax2.set_xlim(0, 1)
        ax2.set_ylim(0, 1)
        ax2.axis('off')
        ax2.set_title('主要数据流')
        
        # 3. 依赖关系图
        ax3 = axes[1, 0]
        dependencies = {
            'bias': ['core'],
            'solvers': ['core', 'bias'],
            'utils': ['core']
        }
        
        positions = {
            'core': (0.5, 0.8),
            'bias': (0.2, 0.5),
            'solvers': (0.8, 0.5),
            'utils': (0.5, 0.2)
        }
        
        # 绘制节点
        for module, pos in positions.items():
            circle = plt.Circle(pos, 0.1, color='lightblue', ec='black', linewidth=2)
            ax3.add_patch(circle)
            ax3.text(pos[0], pos[1], module, ha='center', va='center', fontweight='bold')
        
        # 绘制依赖箭头
        for module, deps in dependencies.items():
            for dep in deps:
                from_pos = positions[dep]
                to_pos = positions[module]
                ax3.arrow(from_pos[0], from_pos[1]-0.1,
                          to_pos[0]-from_pos[0], to_pos[1]-from_pos[1]+0.1,
                          head_width=0.05, head_length=0.05, fc='black')
        
        ax3.set_xlim(0, 1)
        ax3.set_ylim(0, 1)
        ax3.set_aspect('equal')
        ax3.axis('off')
        ax3.set_title('模块依赖关系')
        
        # 4. 控制流图
        ax4 = axes[1, 1]
        for flow_name, steps in self.control_flow.items():
            ax4.text(0.05, 0.9 - list(self.control_flow.keys()).index(flow_name) * 0.3,
                   f"{flow_name}:", fontweight='bold')
            
            for i, step in enumerate(steps[:3]):  # 只显示前3步
                ax4.text(0.05, 0.85 - list(self.control_flow.keys()).index(flow_name) * 0.3 - i * 0.08,
                           f"  {i+1}. {step}", fontsize=8)
        
        ax4.set_xlim(0, 1)
        ax4.set_ylim(0, 1)
        ax4.axis('off')
        ax4.set_title('主要控制流')
        
        plt.tight_layout()
        plt.show()
    
    def print_interaction_analysis(self):
        """打印交互分析报告"""
        print("\n模块交互分析报告")
        print("=" * 50)
        
        print("\n1. 依赖层次：")
        print("   Level 0: core (基础层)")
        print("   Level 1: bias, utils (依赖core)")
        print("   Level 2: solvers (依赖core, bias, utils)")
        
        print("\n2. 松耦合设计：")
        print("   - 通过接口而非实现进行交互")
        print("   - 使用依赖注入减少耦合")
        print("   - 事件驱动架构支持扩展")
        
        print("\n3. 关键交互点：")
        print("   - Problem.evaluate(): 问题评估接口")
        print("   - Bias.apply(): 偏置应用接口")
        print("   - EliteManager.update(): 精英更新接口")
        print("   - Solver.solve(): 统一求解接口")

# 创建交互分析器
interaction_analyzer = ModuleInteractionAnalyzer()

# 打印分析报告
interaction_analyzer.print_interaction_analysis()

# 可视化交互
interaction_analyzer.visualize_interactions()
```

## 4. 扩展机制

展示框架的扩展点和扩展方法。

```python
class ExtensionPoint:
    """扩展点基类"""
    
    def __init__(self, name: str):
        self.name = name
        self.extensions = {}
    
    def register(self, extension_name: str, extension_class: type):
        """注册扩展"""
        self.extensions[extension_name] = extension_class
        print(f"注册扩展: {extension_name} -> {extension_class.__name__}")
    
    def get_extension(self, extension_name: str):
        """获取扩展"""
        return self.extensions.get(extension_name)
    
    def list_extensions(self):
        """列出所有扩展"""
        return list(self.extensions.keys())

# 定义主要的扩展点
class FrameworkExtensions:
    """框架扩展管理器"""
    
    def __init__(self):
        self.extension_points = {
            'solvers': ExtensionPoint('求解器扩展点'),
            'biases': ExtensionPoint('偏置扩展点'),
            'problems': ExtensionPoint('问题扩展点'),
            'evaluators': ExtensionPoint('评估器扩展点'),
            'visualizers': ExtensionPoint('可视化扩展点')
        }
    
    def register_solver(self, name: str, solver_class: type):
        """注册新求解器"""
        # 验证求解器类是否实现必要接口
        required_methods = ['solve', '_initialize', '_step']
        for method in required_methods:
            if not hasattr(solver_class, method):
                raise ValueError(f"求解器类必须实现 {method} 方法")
        
        self.extension_points['solvers'].register(name, solver_class)
    
    def register_bias(self, name: str, bias_class: type):
        """注册新偏置"""
        required_methods = ['apply']
        for method in required_methods:
            if not hasattr(bias_class, method):
                raise ValueError(f"偏置类必须实现 {method} 方法")
        
        self.extension_points['biases'].register(name, bias_class)
    
    def create_solver(self, name: str, **kwargs):
        """创建求解器实例"""
        solver_class = self.extension_points['solvers'].get_extension(name)
        if solver_class is None:
            raise ValueError(f"未知的求解器: {name}")
        
        return solver_class(**kwargs)
    
    def create_bias(self, name: str, **kwargs):
        """创建偏置实例"""
        bias_class = self.extension_points['biases'].get_extension(name)
        if bias_class is None:
            raise ValueError(f"未知的偏置: {name}")
        
        return bias_class(**kwargs)
    
    def list_available(self):
        """列出所有可用的扩展"""
        for point_name, point in self.extension_points.items():
            print(f"\n【{point_name.upper()}】")
            extensions = point.list_extensions()
            if extensions:
                for ext in extensions:
                    print(f"  - {ext}")
            else:
                print("  (暂无扩展)")

# 示例：创建自定义求解器
class CustomParticleSwarmOptimizer:
    """自定义粒子群优化求解器"""
    
    def __init__(self, swarm_size=50, w=0.7, c1=1.5, c2=1.5):
        self.swarm_size = swarm_size
        self.w = w  # 惯性权重
        self.c1 = c1  # 个体学习因子
        self.c2 = c2  # 社会学习因子
        
        self.particles = []
        self.velocities = []
        self.personal_best = []
        self.global_best = None
        self.global_best_fitness = float('inf')
    
    def solve(self, problem, max_evaluations=10000, **kwargs):
        """求解接口"""
        print(f"\n运行粒子群优化: {problem.name}")
        print(f"粒子数: {self.swarm_size}")
        
        # 初始化
        self._initialize(problem)
        
        # 主循环
        evaluations = 0
        while evaluations < max_evaluations:
            self._update_particles(problem)
            evaluations += self.swarm_size
        
        return {
            'best_solution': self.global_best,
            'best_fitness': self.global_best_fitness
        }
    
    def _initialize(self, problem):
        """初始化粒子群"""
        # 初始化粒子和速度
        for _ in range(self.swarm_size):
            particle = np.array([
                np.random.uniform(low, high) 
                for low, high in problem.bounds
            ])
            velocity = np.zeros_like(particle)
            
            self.particles.append(particle)
            self.velocities.append(velocity)
            self.personal_best.append(particle.copy())
    
    def _update_particles(self, problem):
        """更新粒子位置和速度"""
        for i in range(self.swarm_size):
            # 更新速度
            r1, r2 = np.random.random(2)
            self.velocities[i] = (
                self.w * self.velocities[i] +
                self.c1 * r1 * (self.personal_best[i] - self.particles[i]) +
                self.c2 * r2 * (self.global_best - self.particles[i])
            )
            
            # 更新位置
            self.particles[i] += self.velocities[i]
            
            # 评估适应度
            fitness = problem.evaluate(self.particles[i])
            
            # 更新个体最优
            current_fitness = problem.evaluate(self.personal_best[i])
            if fitness < current_fitness:
                self.personal_best[i] = self.particles[i].copy()
            
            # 更新全局最优
            if fitness < self.global_best_fitness:
                self.global_best_fitness = fitness
                self.global_best = self.particles[i].copy()
    
    def _initialize(self, problem):
        pass  # 兼容接口
    
    def _step(self, problem):
        pass  # 兼容接口

# 示例：创建自定义偏置
class MomentumBias:
    """动量偏置 - 保持搜索方向"""
    
    def __init__(self, momentum=0.9, decay_rate=0.99):
        self.momentum = momentum
        self.decay_rate = decay_rate
        self.last_direction = None
    
    def apply(self, population, problem):
        """应用动量偏置"""
        if self.last_direction is None:
            # 第一次，初始化方向
            self.last_direction = np.random.randn(problem.n_var)
        
        biased_population = []
        
        for individual in population:
            # 添加动量偏置
            momentum_term = self.momentum * self.last_direction
            biased_individual = individual + momentum_term
            
            # 边界处理
            for i, (low, high) in enumerate(problem.bounds):
                biased_individual[i] = np.clip(biased_individual[i], low, high)
            
            biased_population.append(biased_individual)
        
        # 更新方向（简化：使用种群平均方向）
        if population:
            mean_direction = np.mean(population, axis=0)
            self.last_direction = (
                self.momentum * self.last_direction + 
                (1 - self.momentum) * mean_direction
            )
            
            # 衰减
            self.momentum *= self.decay_rate
        
        return biased_population

# 使用扩展机制
print("\n框架扩展机制演示：")
print("=" * 50)

# 创建扩展管理器
extensions = FrameworkExtensions()

# 注册自定义求解器
print("\n1. 注册自定义求解器：")
extensions.register_solver('pso', CustomParticleSwarmOptimizer)
extensions.register_solver('pso_adaptive', 
                         lambda: CustomParticleSwarmOptimizer(w=0.8, c1=2.0, c2=2.0))

# 注册自定义偏置
print("\n2. 注册自定义偏置：")
extensions.register_bias('momentum', MomentumBias)

# 列出所有扩展
print("\n3. 可用扩展列表：")
extensions.list_available()

# 使用扩展创建实例
print("\n4. 使用扩展创建实例：")
try:
    # 创建求解器
    pso_solver = extensions.create_solver('pso', swarm_size=30)
    print(f"创建求解器成功: {pso_solver.__class__.__name__}")
    
    # 创建偏置
    momentum_bias = extensions.create_bias('momentum', momentum=0.95)
    print(f"创建偏置成功: {momentum_bias.__class__.__name__}")
except Exception as e:
    print(f"创建实例失败: {e}")

# 展示扩展性原则
print("\n\n扩展性设计原则：")
print("-" * 30)
print("1. 开闭原则：对扩展开放，对修改封闭")
print("2. 接口隔离：定义清晰的扩展接口")
print("3. 依赖倒置：依赖抽象而非具体实现")
print("4. 单一职责：每个扩展点职责明确")
print("5. 组合优于继承：灵活组合不同扩展")
```

```python
class ConfigurationManager:
    """配置管理器"""
    
    def __init__(self):
        self.config = {
            'solver': {
                'default': 'nsga2',
                'population_size': 100,
                'max_generations': 500,
                'crossover_rate': 0.9,
                'mutation_rate': 0.1
            },
            'bias': {
                'enabled': True,
                'system': 'adaptive',
                'strength': 0.5,
                'update_frequency': 10
            },
            'parallel': {
                'enabled': False,
                'n_workers': 4,
                'backend': 'multiprocessing',
                'chunk_size': None
            },
            'memory': {
                'max_cache_size': 1000,
                'gc_threshold': 0.8,
                'auto_gc': True
            },
            'logging': {
                'level': 'INFO',
                'file': None,
                'format': '%(asctime)s - %(name)s - %(levelname)s - %(message)s'
            }
        }
    
    def load_from_file(self, filepath: str):
        """从文件加载配置"""
        try:
            import json
            with open(filepath, 'r') as f:
                file_config = json.load(f)
            
            # 深度合并配置
            self._deep_merge(self.config, file_config)
            print(f"配置已从 {filepath} 加载")
        except Exception as e:
            print(f"加载配置失败: {e}")
    
    def save_to_file(self, filepath: str):
        """保存配置到文件"""
        try:
            import json
            with open(filepath, 'w') as f:
                json.dump(self.config, f, indent=2)
            print(f"配置已保存到 {filepath}")
        except Exception as e:
            print(f"保存配置失败: {e}")
    
    def update(self, section: str, key: str, value):
        """更新配置项"""
        if section in self.config:
            self.config[section][key] = value
            print(f"更新配置: {section}.{key} = {value}")
        else:
            print(f"未知的配置节: {section}")
    
    def get(self, section: str, key: str, default=None):
        """获取配置项"""
        return self.config.get(section, {}).get(key, default)
    
    def _deep_merge(self, base_dict: dict, update_dict: dict):
        """深度合并字典"""
        for key, value in update_dict.items():
            if key in base_dict and isinstance(base_dict[key], dict) and isinstance(value, dict):
                self._deep_merge(base_dict[key], value)
            else:
                base_dict[key] = value
    
    def print_config(self):
        """打印当前配置"""
        import json
        print("\n当前配置：")
        print("=" * 50)
        print(json.dumps(self.config, indent=2, ensure_ascii=False))

# 最佳实践指南
class BestPracticesGuide:
    """最佳实践指南"""
    
    @staticmethod
    def problem_modeling():
        """问题建模最佳实践"""
        practices = [
            "1. 继承合适的基类（Problem/MultiObjectiveProblem）",
            "2. 正确定义变量边界和约束",
            "3. 实现高效的evaluate方法",
            "4. 考虑批量评估优化",
            "5. 提供有意义的名称和文档",
            "6. 处理边界情况和异常",
            "7. 考虑数值稳定性"
        ]
        
        print("\n问题建模最佳实践：")
        for practice in practices:
            print(f"  {practice}")
    
    @staticmethod
    def solver_selection():
        """求解器选择指南"""
        guidelines = {
            "单目标，小规模": [
                "蒙特卡洛",
                "局部搜索"
            ],
            "单目标，大规模": [
                "遗传算法",
                "粒子群优化"
            ],
            "多目标": [
                "NSGA-II（经典选择）",
                "MOEA/D（分解方法）"
            ],
            "昂贵评估": [
                "贝叶斯优化",
                "代理模型辅助"
            ],
            "需要并行": [
                "支持并行评估的算法",
                "岛模型遗传算法"
            ]
        }
        
        print("\n求解器选择指南：")
        for scenario, solvers in guidelines.items():
            print(f"\n{scenario}:")
            for solver in solvers:
                print(f"  • {solver}")
    
    @staticmethod
    def performance_optimization():
        """性能优化建议"""
        tips = [
            "1. 使用向量化操作替代Python循环",
            "2. 启用并行评估（适当的问题）",
            "3. 合理设置种群大小和代数",
            "4. 使用精英管理避免重复评估",
            "5. 批量处理减少函数调用开销",
            "6. 使用缓存机制存储重复计算",
            "7. 监控内存使用，及时优化",
            "8. 选择合适的数据结构",
        ]
        
        print("\n性能优化建议：")
        for tip in tips:
            print(f"  {tip}")
    
    @staticmethod
    def debugging_tips():
        """调试技巧"""
        tips = [
            "1. 使用小规模问题验证算法",
            "2. 启用详细日志记录",
            "3. 可视化优化过程",
            "4. 检查边界条件",
            "5. 验证已知基准问题",
            "6. 使用断点调试复杂逻辑",
            "7. 分析收敛历史",
            "8. 对比不同参数设置",
        ]
        
        print("\n调试技巧：")
        for tip in tips:
            print(f"  {tip}")

# 创建配置管理器
config_manager = ConfigurationManager()

# 展示配置管理
print("配置管理演示：")
print("=" * 50)

# 更新配置
config_manager.update('solver', 'population_size', 150)
config_manager.update('parallel', 'enabled', True)
config_manager.update('bias', 'strength', 0.7)

# 读取配置
print(f"\n读取配置：")
print(f"种群大小: {config_manager.get('solver', 'population_size')}")
print(f"并行启用: {config_manager.get('parallel', 'enabled')}")
print(f"偏置强度: {config_manager.get('bias', 'strength')}")

# 创建示例配置文件
example_config = {
    "solver": {
        "default": "nsga2",
        "max_generations": 1000
    },
    "bias": {
        "enabled": True,
        "strength": 0.6
    },
    "custom_setting": {
        "my_parameter": 42
    }
}

# 保存配置到文件
config_file = 'example_config.json'
with open(config_file, 'w') as f:
    import json
    json.dump(example_config, f, indent=2)

# 从文件加载配置
config_manager.load_from_file(config_file)

print(f"\n从文件加载后：")
print(f"最大代数: {config_manager.get('solver', 'max_generations')}")
print(f"自定义参数: {config_manager.get('custom_setting', 'my_parameter')}")

# 清理文件
os.remove(config_file)

# 展示最佳实践
BestPracticesGuide.problem_modeling()
BestPracticesGuide.solver_selection()
BestPracticesGuide.performance_optimization()
BestPracticesGuide.debugging_tips()

# 总结架构要点
print("\n\nNSGABLACK框架架构要点总结：")
print("=" * 50)
print("\n核心设计原则：")
print("1. 模块化：清晰的职责划分")
print("2. 可扩展：插件式架构设计")
print("3. 松耦合：接口驱动交互")
print("4. 高性能：并行和优化支持")

print("\n主要优势：")
print("• 统一的问题和求解器接口")
print("• 灵活的偏置系统")
print("• 丰富的算法库")
print("• 完善的工具支持")
print("• 易于扩展和定制")

print("\n适用场景：")
print("• 多目标优化问题")
print("• 大规模复杂优化")
print("• 算法研究和比较")
print("• 工程优化应用")
print("• 教学和演示")
```
