# NSGABLACK Bias System Core Concepts

## What is the Bias System?

The Bias System is the core innovation of NSGABLACK, responsible for guiding the optimization process by injecting prior knowledge and learning information. It's not about prejudice, but about **intelligent guidance**.

### Key Roles
1. **Search Direction Control**: Guides the algorithm toward promising regions
2. **Knowledge Integration**: Incorporates domain expertise and historical experience
3. **Adaptive Adjustment**: Dynamically adjusts search strategies

## Core Components of the Bias System

### 1. Algorithmic Bias
Affects genetic operations within the algorithm:
- Crossover probability adjustment
- Mutation strategy selection
- Selection pressure control

### 2. Domain Knowledge Bias
Utilizes problem-specific information:
- Feasible region constraints
- Heuristic rules
- Historical optimal solutions

### 3. Adaptive Bias
Adjusts based on search dynamics:
- Convergence state
- Diversity metrics
- Search history

## Simple Bias Implementation Example

Let's create a simple bias system to understand how it works:

````python
class SimpleBias:
    def __init__(self, bias_strength=0.1):
        self.bias_strength = bias_strength
        self.generation = 0
        
    def apply(self, population, problem, generation):
        """Apply bias to the population
        
        Args:
            population: Current population (list of individuals)
            problem: Optimization problem
            generation: Current generation number
        
        Returns:
            Biased population
        """
        self.generation = generation
        biased_population = []
        
        for individual in population:
            # Simulate bias effect: gradually converge toward optimal values
            biased_individual = self._apply_bias(individual)
            biased_population.append(biased_individual)
            
        return biased_population
    
    def _apply_bias(self, individual):
        """Apply bias to a single individual"""
        # Simple example: bias toward lower values
        bias_factor = 1 - self.bias_strength * (self.generation / 100)
        
        if isinstance(individual, np.ndarray):
            return individual * bias_factor
        else:
            return individual

# Test the simple bias system
test_pop = [np.random.rand(10) for _ in range(5)]
simple_bias = SimpleBias(bias_strength=0.2)

print("Original population average:", np.mean([np.mean(ind) for ind in test_pop]))

biased_pop = simple_bias.apply(test_pop, None, 10)
print("Biased population average:", np.mean([np.mean(ind) for ind in biased_pop]))
````

## Visualizing Bias Effects

Let's visualize how the bias system affects the search process:

````python
# Visualize bias effects
generations = 50
pop_size = 20
dim = 2

# Store optimization data
no_bias_history = []
with_bias_history = []

# Run without bias
pop_no_bias = [np.random.randn(dim) for _ in range(pop_size)]
for gen in range(generations):
    # Simulate evolution: random search
    pop_no_bias = [ind + np.random.randn(dim) * 0.1 for ind in pop_no_bias]
    best_fitness = min([np.sum(ind**2) for ind in pop_no_bias])  # Sphere function
    no_bias_history.append(best_fitness)

# Run with bias
bias = SimpleBias(bias_strength=0.3)
pop_with_bias = [np.random.randn(dim) for _ in range(pop_size)]
for gen in range(generations):
    # Apply bias
    pop_with_bias = bias.apply(pop_with_bias, None, gen)
    # Simulate evolution
    pop_with_bias = [ind + np.random.randn(dim) * 0.1 for ind in pop_with_bias]
    best_fitness = min([np.sum(ind**2) for ind in pop_with_bias])
    with_bias_history.append(best_fitness)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(no_bias_history, 'b-', label='Without Bias', linewidth=2)
plt.plot(with_bias_history, 'r-', label='With Bias', linewidth=2)
plt.xlabel('Generation')
plt.ylabel('Best Fitness')
plt.title('Effect of Bias on Optimization')
plt.legend()
plt.grid(True)
plt.show()
````

### 3. Dynamic Adjustment
The bias system can automatically adjust based on optimization state:

````python
class AdaptiveBiasSystem:
    def __init__(self):
        self.bias_strength = 0.1
        self.diversity_threshold = 0.5
        self.convergence_rate = 0.95
        
    def update_bias(self, population, generation):
        """Dynamically adjust bias based on population state"""
        # Calculate population diversity
        diversity = self._calculate_diversity(population)
        
        # Adjust bias strength
        if diversity < self.diversity_threshold:
            # Low diversity: reduce bias
            self.bias_strength *= 0.9
        else:
            # High diversity: can increase bias
            self.bias_strength = min(0.3, self.bias_strength * 1.1)
            
        return self.bias_strength
    
    def _calculate_diversity(self, population):
        """Calculate population diversity metric"""
        if len(population) < 2:
            return 1.0
        
        # Simple diversity metric: mean distance
        distances = []
        for i in range(len(population)):
            for j in range(i+1, len(population)):
                dist = np.linalg.norm(population[i] - population[j])
                distances.append(dist)
                
        return np.mean(distances) if distances else 0

# Test adaptive bias system
adaptive_bias = AdaptiveBiasSystem()
test_pop = [np.random.rand(5) for _ in range(10)]

bias_history = []
for gen in range(30):
    bias = adaptive_bias.update_bias(test_pop, gen)
    bias_history.append(bias)
    # Simulate some population changes
    test_pop = [ind + np.random.randn(5) * 0.05 for ind in test_pop]

plt.figure(figsize=(10, 4))
plt.plot(bias_history, 'g-', linewidth=2)
plt.xlabel('Generation')
plt.ylabel('Bias Strength')
plt.title('Adaptive Bias Adjustment')
plt.grid(True)
plt.show()
````

## Best Practices

1. **Bias Strength Control**: Too much bias may lead to premature convergence
2. **Dynamic Adjustment**: Adjust bias based on search state
3. **Multiple Bias Strategies**: Combine different types of bias
4. **Problem Specificity**: Design bias for specific problem characteristics
